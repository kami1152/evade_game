```python
import random
import operator
import copy
```

```python
class pos:
  def __init__(self):
    self.frag = 0
    self.pre_x = 0
    self.pre_y = 0
    self._dict={}
    self.Map =[]
    self.row = 0
    self.col = 0
    self.x_pos = 0
    self.y_pos = 0
    self.cal_x = 0
    self.cal_y = 0
    self.Player = []
    self.Monster1 = []
    self.Monster2 = []
    self.Monster3 = []
    self.temp = []
    self.min_Monster = []
    return

  def init_Field(self):
    while True:
      print('필드의 크기를 정해주세요\n')
      print('가로의 크기 : ')
      try:
        m = int(input())
        if type(m) == int:
          print(f'가로의 크기는 :{m} \n')
          print('세로의 크기를 정해주세요 :')
          n = int(input())
          if type(n) == int:
            print(f'세로의 크기는 :{n}')
            self.row = m
            self.col = n
            self.Map=[[0 for i in range(self.row)] for j in range(self.col)]
            break
      except:
        print('숫자만 입력하세요')

  def spon(self): # 4,4 4,4
    while True:
      for i in range(4):
        self.x_pos = random.randrange(0,self.row) #0 1 2 3 4 5
        self.y_pos = random.randrange(0,self.col) # 0 1 2 3 4 5 6 7
        if i == 0:
          self.Monster1 = [self.x_pos , self.y_pos] #[ 1 , 2]
        elif i == 1:
          self.Monster2 = [self.x_pos , self.y_pos]
        elif i == 2:
          self.Monster3 = [self.x_pos , self.y_pos]
        elif i == 3:
          self.Player = [self.x_pos, self.y_pos]
      if self.Monster1 != self.Monster2 and self.Monster1 != self.Monster3 and self.Monster2 != self.Monster3:
        if self.Player != self.Monster1 and self.Player != self.Monster2 and self.Player != self.Monster3:
          break
    self.Map[self.Monster1[1]][self.Monster1[0]] = 11
    self.Map[self.Monster2[1]][self.Monster2[0]] = 22
    self.Map[self.Monster3[1]][self.Monster3[0]] = 33
    self.Map[self.Player[1]][self.Player[0]] = 55

  def P_length_cal(self): #P = [1,0]
    P_M1 = abs(self.Player[0] - self.Monster1[0]) + abs(self.Player[1] - self.Monster1[1])
    P_M2 = abs(self.Player[0] - self.Monster2[0]) + abs(self.Player[1] - self.Monster2[1])
    P_M3 = abs(self.Player[0] - self.Monster3[0]) + abs(self.Player[1] - self.Monster3[1])
    self._dict = {P_M1:self.Monster1, P_M2:self.Monster2, P_M3:self.Monster3}
    self.min_Monster = min(self._dict.items(), key=operator.itemgetter(0))[1] #{거리 : monseter}
    
  def monster1_move(self):
    self.temp = copy.deepcopy(self.Monster1) #temp = 몬스터1 좌표
    r_p = self.Monster1[0] - self.Player[0] #M = 0  P = 1
    c_p = self.Monster1[1] - self.Player[1] #M = 1  P = 3
    if r_p > 0 and self.Monster1[0] > 0:
      self.Monster1[0] = self.Monster1[0]-1
    elif r_p < 0 and self.Monster1[0] < self.row-1:
      self.Monster1[0] = self.Monster1[0]+1
    else:
      if c_p > 0 and self.Monster1[1] > 0:
        self.Monster1[1] = self.Monster1[1]-1
      elif c_p < 0 and self.Monster1[1] < self.col-1:
        self.Monster1[1] = self.Monster1[1]+1
      else:
        self.Monster1[1] = self.Monster1[1] 
    if self.Monster1 != self.Monster2 and self.Monster1 != self.Monster3:
      self.Map[self.Monster1[1]][self.Monster1[0]] = 11 #[2,4]
      self.Map[self.temp[1]][self.temp[0]] = 1 
      print('성공')
    else:
      self.Monster1 = self.temp
      print('실패')
    self.temp = []

  def monster2_move(self):
    self.temp = copy.deepcopy(self.Monster2) #temp = 몬스터1 좌표
    r_p = self.Monster2[0] - self.Player[0] #M = 0  P = 1
    c_p = self.Monster2[1] - self.Player[1] #M = 1  P = 3
    if r_p > 0 and self.Monster2[0] > 0:
      self.Monster2[0] = self.Monster2[0]-1
    elif r_p < 0 and self.Monster2[0] < self.row-1:
      self.Monster2[0] = self.Monster2[0]+1
    else:
      if c_p > 0 and self.Monster2[1] > 0:
        self.Monster2[1] = self.Monster2[1]-1
      elif c_p < 0 and self.Monster2[1] < self.col-1:
        self.Monster2[1] = self.Monster2[1]+1
      else:
        self.Monster2[1] = self.Monster2[1] 
    if self.Monster1 != self.Monster2 and self.Monster2 != self.Monster3:
      self.Map[self.Monster2[1]][self.Monster2[0]] = 22 #[2,4]
      self.Map[self.temp[1]][self.temp[0]] = 2 
      print('성공')
    else:
      self.Monster2 = self.temp
      print('실패')
    self.temp = []

  def monster3_move(self):
    self.temp = copy.deepcopy(self.Monster3) #temp = 몬스터1 좌표
    r_p = self.Monster3[0] - self.Player[0] #M = 0  P = 1
    c_p = self.Monster3[1] - self.Player[1] #M = 1  P = 3
    if c_p > 0 and self.Monster3[1] > 0:
      self.Monster3[1] = self.Monster3[1]-1
    elif c_p < 0 and self.Monster3[1] < self.col-1:
      self.Monster3[1] = self.Monster3[1]+1
    else:
      if r_p > 0 and self.Monster3[0] > 0:
        self.Monster3[0] = self.Monster3[0]-1
      elif r_p < 0 and self.Monster3[0] < self.row-1:
        self.Monster3[0] = self.Monster3[0]+1
      else:
        self.Monster3[0] = self.Monster3[0]
  
    # if r_p > 0 and self.Monster3[0] > 0:
    #   self.Monster3[0] = self.Monster3[0]-1
    # elif r_p < 0 and self.Monster3[0] < self.row-1:
    #   self.Monster3[0] = self.Monster3[0]+1
    # else:
    #   if c_p > 0 and self.Monster3[1] > 0:
    #     self.Monster3[1] = self.Monster3[1]-1
    #   elif c_p < 0 and self.Monster3[1] < self.col-1:
    #     self.Monster3[1] = self.Monster3[1]+1
    #   else:
    #     self.Monster3[1] = self.Monster3[1] 

    if self.Monster1 != self.Monster3 and self.Monster2 != self.Monster3:
      self.Map[self.Monster3[1]][self.Monster3[0]] = 33 #[2,4]
      self.Map[self.temp[1]][self.temp[0]] = 3
      print('성공')
    else:
      self.Monster3 = self.temp
      print('실패')
    self.temp = []

  def Player_move(self):
    if self.Player == self.Monster1 or self.Player == self.Monster2 or self.Player == self.Monster3:
      print('게임 오버')
      return
    cnt = 0
    self.temp = copy.deepcopy(self.Player)
    r_p = self.Player[0] - self.min_Monster[0] #P = 2  min = 4
    c_p = self.Player[1] - self.min_Monster[1]
    self.Map[self.Player[1]][self.Player[0]] = 1
    if r_p < 0 and self.Player[0] > 0:
      self.Player[0] = self.Player[0]-1
    elif r_p > 0 and self.Player[0] < self.row-1:
      self.Player[0] = self.Player[0]+1
    else:
      frag = random.randrange(1,4)
      if frag == 1:
        if self.Player[0] > 0:
          self.Player[0] = self.Player[0]-1
          cnt += 1
      elif frag == 2:
        if self.Player[0] < self.row-1:
          self.Player[0] = self.Player[0]+1
          cnt += 1
      elif frag == 3:
          self.Player[0] = self.Player[0]
      print(frag)
      frag = 0

    if c_p < 0 and self.Player[1] > 0:
      self.Player[1] = self.Player[1]-1
    elif c_p > 0 and self.Player[1] < self.col-1:
      self.Player[1] = self.Player[1]+1
    else:
      if cnt != 1:
        frag = random.randrange(1,4)
        if frag == 1:
          if self.Player[1] > 0:
            self.Player[1] = self.Player[1]-1
        if frag == 2:
          if self.Player[1] < self.col-1:
            self.Player[1]=self.Player[1]+1
        else:
          self.Player[1] = self.Player[1]

        frag = 0
      else:
        self.Player[1] = self.Player[1]

    if self.Player != self.Monster1 and self.Player != self.Monster2 and self.Player != self.Monster3:
      self.Map[self.temp[1]][self.temp[0]] = 5
      self.Map[self.Player[1]][self.Player[0]] = 55
    else:
      self.Player = self.temp
    self.temp = []
```
